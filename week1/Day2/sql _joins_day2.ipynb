```python
%sql

CREATE TABLE IF NOT EXISTS employees ( 
    emp_id INT PRIMARY KEY, 
    emp_name VARCHAR(50), 
    manager_id INT, 
    dept_id INT 
);

CREATE TABLE IF NOT EXISTS departments ( 
    dept_id INT PRIMARY KEY, 
    dept_name VARCHAR(50) 
);

CREATE TABLE IF NOT EXISTS projects ( 
    project_id INT PRIMARY KEY, 
    project_name VARCHAR(50), 
    emp_id INT 
);
```

```python
%sql

CREATE OR REPLACE TABLE employees AS
SELECT * FROM VALUES
(1, 'Karthik', NULL, 1), 
(2, 'Ajay', 1, 1), 
(3, 'Vijay', 1, 2), 
(4, 'Vinay', 2, 2), 
(5, 'Meena', 3, 3), 
(6, 'Veer', NULL, 4), 
(7, 'Keerthi', 4, 5), 
(8, 'Priya', 4, 5)
AS t(emp_id, emp_name, manager_id, dept_id);

CREATE OR REPLACE TABLE departments AS
SELECT * FROM VALUES
(1, 'HR'), 
(2, 'IT'), 
(3, 'Finance'), 
(4, 'Marketing'), 
(5, 'Sales')
AS t(dept_id, dept_name);

CREATE OR REPLACE TABLE projects AS
SELECT * FROM VALUES
(1, 'Project A', 1), 
(2, 'Project B', 2), 
(3, 'Project C', 3), 
(4, 'Project D', 4), 
(5, 'Project E', 5)
AS t(project_id, project_name, emp_id);
```

### Retrieve the names of employees and their corresponding managers from the "employees" table, ensuring that even employees without managers are included

```python
%sql
SELECT e.emp_name AS employee_name,
       m.emp_name AS manager_name
FROM employees e
LEFT JOIN employees m
ON e.manager_id = m.emp_id;
```

### Display all employees and their corresponding departments from the "employees" and "departments" tables, showing employees even if they don't belong to any department. 


```python
%sql
SELECT e.emp_name,d.dept_name
FROM employees e
LEFT JOIN departments d   
ON e.dept_id = d.dept_id;
```

### List the names of employees who report to a manager, along with their manager's name, from the "employees" table

```python
%sql
SELECT e.emp_name AS employee_name,
       m.emp_name AS manager_name
FROM employees e
JOIN employees m
ON e.manager_id = m.emp_id;
    
```

### Find the total number of employees who belong to a department, ensuring the departments with no employees are still included.

```python
%sql
SELECT d.dept_name, COUNT(e.emp_id) AS emp_count
FROM departments d
LEFT JOIN employees e ON d.dept_id = e.dept_id
GROUP BY d.dept_name;
```
